```python
import numpy as np
```

```python
# 공부한 시간과 점수를 각각 x.y라는 이름의 넘파이 배열로 만듦
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])
```

```python
# x의 평균값, y의 평균값을 구함
mx = np.mean(x)
my = np.mean(y)
```

```python
# 기울기 공식의 분모 부분
divisor = sum([(i - mx)**2 for i in x])
```

```python
# 기울기 공식의 분자 부분
def top(x, mx, y, my):
  d = 0
  for i in range(len(x)):
    d += (x[i]-mx)*(y[i]-my)
  return d
dividend = top(x, mx, y, my)
```

```python
# 기울기 a를 구하는 공식과 y절편 b를 구하는 공식
a = dividend/divisor
b = my - (mx*a)
```

```python
print("x의 평군값: ", mx)
print("y의 평군값: ", my)
print()
print("분모: ", divisor)
print("분자: ", dividend)
print()
print("기울기 a = ", a)
print("기울기 b = ", b)
```

```python
# 4-6 파이썬 코딩으로 확인하는 평균 제곱 오차
```

```python
# 임의로 기울기 fake_a와 y절편 fake_b를 정함
fake_a = 3
fake_b = 76
```

```python
# 기울기가 fake_a이고 y절편이 fake_b인 함수식
def predict(x):
  return fake_a * x + fake_b
```

```python
# 위 코드의 결괏값이 들어갈 빈 리스트를 만듦
predict_result = []

# 모든 x값을 preodict() 함수에 한번씩 대입해 예측값 리스트를 채우는 코드 작성
for i in range(len(x)):
  predict_result.append(predict(x[i]))
  print("공부시간=%.f, 실제점수=%.f, 예측점수=%.f" % (x[i], y[i], predict(x[i])))
```

```python
# 평균 제곱 오차를 구하는 함수를 만듦
n = len(x)
def mse(y, y_pred):
  return (1/n) * sum((y - y_pred)**2)

print("평균 제곱 오차: " + str(mse(y,predict_result)))
```
