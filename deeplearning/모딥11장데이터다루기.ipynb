```python
# 11-5 피마 인디언의 당뇨병 예측 실행
```

```python
X = df.iloc[:,0:8] # 세부 정보를 X로 지정
y = df.iloc[:,8] # 당뇨병 여부를 y로 지정
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 모델을 설정
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu', name='Dense_1'))
model.add(Dense(8, activation='relu', name='Dense_2'))
model.add(Dense(1, activation='sigmoid', name='Dense_3'))
model.summary()
# Layer : 층의 이름과 유형을 나타냄
# Output Shape : 각 층에 몇 개의 출력이 발생하는지 나타냄
#                쉼표 앞은 행(샘플)의 수, 뒤는 열(속성)의 수를 의미
#                행의 수는 batch_size에 정한 만큼 이별되므로 딥러닝 모델이서는 이를 특별히 세지 않음 -> None
# Param : 파라미터수, 즉 총 가중치와 바이어스 수의 합
# Total params, Trainable params,Non-trainable params
# : 전체 파라미터를 합산한 값.  Trainable params는 학습을 진행하면서 업데이트가 된 파라미터들.
#   Non-trainable params는 업데이트가 되지 않은 파리미터 수    
```

```python
# 모델을 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 모델을 실행
history = model.fit(X,y,epochs=100, batch_size=5)
```
